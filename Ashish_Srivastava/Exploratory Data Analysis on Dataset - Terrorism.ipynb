# Visualising Data

```python
plt.figure(figsize=(30,20))
sns.boxplot(x="Year", data=dfeda, palette='Accent')
plt.xticks(rotation=60)
```

```python
sns.countplot(data=dfeda, x='Month',palette='crest')
```

```python
sns.histplot(x='Year',y='Region',data=dfeda,palette='rainbow')
```

```python
pd.crosstab(dfeda.Year, dfeda.Country).plot(kind='bar',figsize=(80,45),color='brown')
plt.title('Activity of Terrorists each year')
plt.ylabel('No of Attacks')
plt.show()
```

```python
sns.histplot(x='Region', data=dfeda, palette='rainbow')
```

```python
plt.figure(figsize=[30,16])
sns.heatmap(dfeda.corr(),cmap='cividis', linewidths=0.6)
plt.show()
```

```python
plt.figure(figsize = (30,12))
sns.barplot(x=dfeda['Country'].value_counts()[:20].index,y=dfeda['Country'].value_counts()[:20].values)
plt.title('Top 20 Countries Affected')
plt.xlabel('Countries')
plt.ylabel('Attacks')
plt.xticks(rotation=80)
plt.show()
```

```python
plt.figure(figsize = (15,8))
sns.histplot(x=dfeda['state'].value_counts()[:20].index,y=dfeda['state'].value_counts()[:20].values)
plt.title('Top 20 States Affected')
plt.xlabel('States')
plt.ylabel('count')
plt.xticks(rotation=60)
plt.show()
```

```python
plt.figure(figsize = (15,8))
sns.scatterplot(x=dfeda['Region'].value_counts()[:20].index,y=dfeda['Region'].value_counts()[:20].values)
plt.title('Top 20 Region Affected')
plt.xlabel('Region')
plt.ylabel('count')
plt.xticks(rotation=60)
plt.show()
```

```python
plt.figure(figsize = (15,8))
sns.boxplot(x=dfeda['Group'].value_counts()[:20].index,y=dfeda['Group'].value_counts()[:20].values)
plt.title('Top 20 Terrorist Group Affected')
plt.xlabel('Group')
plt.ylabel('count')
plt.xticks(rotation=50)
plt.show()
```

```python
plt.figure(figsize = (15,8))
sns.barplot(x=dfeda['AttackType'].value_counts()[:20].index,y=dfeda['AttackType'].value_counts()[:20].values)
plt.title('Top 20 Types of Attacks Type')
plt.xlabel('AttackType')
plt.ylabel('count')
plt.xticks(rotation=60)
plt.show()
```
